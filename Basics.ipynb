# Let's start on the basics. Variables and arrays

### First, import some libraries

```
import numpy as np
import random
```

```
array_1 = np.array([8,7,5,9,2,10]) # Create an array of 6 values
print (type(array_1), array_1.dtype)
```

```
array_2 = np.array(np.random.randint(0,10,len(array_1))) # Create an array of random int values
                                                      # with the same lenght as array_1

print (array_1, array_2) # print both array_1 and array_2
print (array_1.shape, array_2.shape) # print both array_1 and array_2 shapes
print (array_1 + array_2) # print summation of array_1 and array_2
```

```
array_1[3] = 70 # Replace the value of array_1 in pos 3 by 70
print (array_1 + array_2) # Print summation again
```

```
# array_1[1] = 32 # Replace the values of array_1 in pos 1-3 by 32
# array_1[2] = 32
# array_1[3] = 32
array_1[1:3] = 32
print (array_1 + array_2) # Print summation again
```

```
array_3 = array_1 * array_2 # Apply element-wise multiplication among array_1 and array_2
# [ 8 32 32 32  2 10] [5 5 5 3 7 4]
array_4 = np.dot(array_1, array_2)
print (array_3, array_4)# Print results
```

### Load an image in grayscale

```
# Loading a color image in grayscale
img_gray = cv2.imread( "/content/lena.jpg", 0)
plt.imshow(img_gray, cmap="gray")
```

### Play with the image


```
# Printing the image shape
print (img_gray.shape)
```

```
# Regarding through the image... Again
print (img_gray[0:10, 10:20])

```

```
lena_face = img_gray[100:200, 180:280]
plt.imshow(lena_face, cmap='gray')
```

```
img_gray[200:300, 0:100] = lena_face
plt.imshow(img_gray, cmap='gray')
```

### Going further

```
# Messing everything around.
img = cv2.imread('lena.jpg', 0)
img[0:150, 0:50] = 0
plt.imshow(img, cmap='gray')
```

```
# Playing with image intensities
# Cargar una imagen en escala de grises
img_gray = cv2.imread('lena.jpg', 0)

# Recorrer todos los píxeles de la imagen
for i in range(img_gray.shape[0]) :
  for j in range(img_gray.shape[1]) :
    # Verificar si el valor de intensidad del píxel es mayor que 128
    if img_gray[i, j] > 128:
      # Si es mayor, establecer el valor del píxel en 255 (blanco)
      img_gray[i, j] = 255

# Mostrar la imagen resultante
plt.imshow(img_gray, cmap='gray')
#plt.imshow(img_gray, cmap='red')
```

### Save an image using OpenCV

```
cv2.imwrite('newlena.png',img_gray)
```
